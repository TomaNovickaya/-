```python
import pandas as pd
import random
import numpy as np
from sklearn.preprocessing import OneHotEncoder

lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
#Вариант 1
data['tmp'] = 1
data.set_index([data.index, 'whoAmI'], inplace=True) # в качестве индекса используем столбец whoAmI
data = data.unstack(level=-1, fill_value = 0).astype(int) # поворот столбца, котрый является индексом
data.columns = data.columns.droplevel() # возвращаем столбцы с требуемым индексом
data.columns.name = None
data.head()

#Вариант 2
# data['whoAmI'] = data['whoAmI'].astype('category')
# data['whoAmI_new'] = data['whoAmI'].cat.codes
# enc = OneHotEncoder()
# enc_data = pd.DataFrame(enc.fit_transform(
#     data[['whoAmI_new']]).toarray())
# New_df = data.join(enc_data)
# New_df.head()
```
